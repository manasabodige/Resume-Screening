```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, accuracy_score

data = {
    'text': [
        'Free entry in 2 a wkly comp to win FA Cup final tkts 21st May 2005.',
        'Hi, precious! How are you loving the new job?',
        'WINNER!! As a valued network customer you have been selected to receivea £900 prize reward!',
        'Had your mobile 11 months or more? U R entitled to Update to the latest colour camera mobiles for Free!',
        'I am fine. out for dinner with friends. see you later.',
        'Urgent! Please call 09061743386 from landline. Your complimentary 4* luxury holiday awaits.',
        'Hey there, are we still meeting for lunch today?',
        'Can you send me the lecture notes from yesterday?',
        'PRIVATE! Your 2004 Account Statement shows 800 un-claimed points.',
        'Dear voucher holder, to claim your 1st class airport lounge pass text PLAY to 80101.'
    ] * 20,
    'label': ['spam', 'ham', 'spam', 'spam', 'ham', 'spam', 'ham', 'ham', 'spam', 'spam'] * 20
}

df = pd.DataFrame(data)

X_train, X_test, y_train, y_test = train_test_split(df['text'], df['label'], test_size=0.2, random_state=42)

vectorizer = CountVectorizer(stop_words='english')
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

model = MultinomialNB()
model.fit(X_train_vec, y_train)

y_pred = model.predict(X_test_vec)
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
