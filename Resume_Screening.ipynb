```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score

data = {
    'resume': [
        'Python Developer with 3 years experience building web applications using Django and Flask. Proficient in SQL and AWS cloud technologies.',
        'Data Scientist skilled in machine learning, deep learning, Python, R, and data visualization tools like Tableau and PowerBI.',
        'Human Resources professional managing recruitment, employee onboarding, payroll, and conflict resolution across large organizations.',
        'Java Engineer specializing in microservices architecture, Spring Boot, Hibernate, and continuous integration pipelines using Jenkins.',
        'Digital Marketing specialist experienced in SEO, Google Analytics, social media ad campaigns, content creation, and brand strategy.',
        'Python Backend Developer managing scalable APIs, database optimization, PostgreSQL, Docker containers, and test-driven development.',
        'Data Analyst interpreting complex datasets, cleaning data using Pandas, performing statistical analysis, and generating executive reports.',
        'HR Manager coordinating talent acquisition, employee engagement programs, policy drafting, and performance management systems.',
        'Senior Java Developer designing enterprise applications, cloud infrastructure, Oracle databases, and mentoring junior engineers.',
        'Marketing Manager designing growth marketing funnels, email automation, market research, and multi-channel campaign deployment.'
    ] * 20,
    'category': ['Python', 'Data Science', 'HR', 'Java', 'Marketing', 'Python', 'Data Science', 'HR', 'Java', 'Marketing'] * 20
}

df = pd.DataFrame(data)

X_train, X_test, y_train, y_test = train_test_split(df['resume'], df['category'], test_size=0.2, random_state=42)

vectorizer = TfidfVectorizer(stop_words='english')
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train_vec, y_train)

y_pred = model.predict(X_test_vec)
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
