```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

data = {
    'review': [
        'This movie was absolutely amazing! I loved the plot and the acting was superb.',
        'Worst experience ever. The film was boring, slow, and a total waste of time.',
        'Highly recommended! A brilliant masterpiece with great character development.',
        'I hated this film. It made no sense and the audio quality was terrible.',
        'Truly spectacular cinematography and an outstanding performance by the lead actor.',
        'Extremely disappointing. The storyline was weak and predictable.',
        'An excellent watch for the entire family. Entertaining from start to finish.',
        'Do not watch this. It is incredibly frustrating and dull.',
        'What a beautiful and heartwarming story. I thoroughly enjoyed it.',
        'A complete disaster. Terrible directing and awful visual effects.'
    ] * 20,
    'sentiment': ['positive', 'negative', 'positive', 'negative', 'positive', 'negative', 'positive', 'negative', 'positive', 'negative'] * 20
}

df = pd.DataFrame(data)

X_train, X_test, y_train, y_test = train_test_split(df['review'], df['sentiment'], test_size=0.2, random_state=42)

vectorizer = TfidfVectorizer(stop_words='english')
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

model = LogisticRegression()
model.fit(X_train_vec, y_train)

y_pred = model.predict(X_test_vec)
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
